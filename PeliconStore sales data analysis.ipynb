## exploratory data analysis

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

from th e above graph we say that maximum buyers are female.

```python
sales_gen = df.groupby (['Gender'], as_index = False) ['Sales'].sum().sort_values(by = 'Sales', ascending = False)
sns.barplot(x = 'Gender', y = 'Sales', data = sales_gen)
```

# Method of Payment

```python
ax = sns.countplot(data = df, x = 'Method of Payment', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen2 = df.groupby (['Method of Payment'], as_index = False) ['Sales'].sum().sort_values(by = 'Sales', ascending = False)
sns.barplot(x = 'Method of Payment', y = 'Sales', data = sales_gen2)
```

```python
ax = sns.countplot(data = df, x = 'Marital Status', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen2 = df.groupby (['Marital Status'], as_index = False) ['Sales'].sum().sort_values(by = 'Sales', ascending = False)
sns.barplot(x = 'Marital Status', y = 'Sales', data = sales_gen2)
```

```python
sales_item = df.groupby (['Method of Payment'], as_index = False) ['Items'].sum().sort_values(by = 'Items', ascending = False).head(20)
sns.set(rc={'figure.figsize' : (15,5)})
sns.barplot(data = sales_item, x = 'Method of Payment', y = 'Items')
```
